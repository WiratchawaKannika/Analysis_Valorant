# Import Data

```python
# Read excel file with sheet name
sheetname = ['ลำดับ']

dict_df = pd.read_excel('/home/kannika/Analysis_Valorant/Datamining_model-300.xlsx', 
                   sheet_name=sheetname)
df_sheet = dict_df['ลำดับ']
df_sheet = df_sheet.iloc[:, 0:4, ]
print(df_sheet.shape)
df_sheet
```

- (1) Cut off Value "-" in Action 3

- (2) Cut off "Buy_Aciton"

- (3) Cut off "Class 0 Aciton"

```python
## - 1
df_sheet2 = df_sheet.loc[df_sheet["Action 3"] != "-"].reset_index(drop=True)
#onvert to string ==> ["ลำดับ"]  
#df_sheet2['ลำดับ'] = df_sheet2['ลำดับ'].apply(str)
print(df_sheet2.shape)
df_sheet2
```

```python
# print(df_sheet2['ลำดับ'][0])
# print(type(df_sheet2['ลำดับ'][0]))
```

```python
## --2 & 3
order_act, index_ = [], []
columns_action = ['Action 1','Action 2','Action 3']
for i in range(len(df_sheet2)):
    for j in range(len(columns_action)):
        value_i = df_sheet2[columns_action[j]][i]
        if value_i == "Buy_Aciton" or value_i == "Class 0 Aciton":
            index_.append(i)
            order_act.append(df_sheet2['ลำดับ'][i])
```

```python
set_order_act = list(set(order_act))
#set_order_act.sort()
set_index_ = list(set(index_))

print(len(set_order_act))
print(len(set_index_))
```
