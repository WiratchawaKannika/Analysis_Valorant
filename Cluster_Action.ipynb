```python
X = dfAll_Stream.iloc[:,7:591]
y = dfAll_Stream.iloc[:,592]

print(X.shape)
print(y.shape)
```

```python
x_scaled = StandardScaler().fit_transform(X)
x_scaled
```

```python
# Dimention Reduction
pca = PCA(n_components=2)
pca_features = pca.fit_transform(x_scaled)
 
# Show PCA characteristics
print('Shape before PCA: ', x_scaled.shape)
print('Shape after PCA: ', pca_features.shape)

print ('Cumulative variance explained by 2 principal components: {:.2%}'.format(np.sum(pca.explained_variance_ratio_)))
```

```python
#Create PCA DataFrame 
pca_df = pd.DataFrame(data=pca_features, columns=['Principal Component 1', 'Principal Component 2'])
print(pca_df.shape)
pca_df
```

```python
# Map target names to targets
pca_df['target'] = dfAll_Stream['sheet']
pca_df.sample(10)
```
